```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.preprocessing import StandardScaler
```

```python
#:Load the data
file_name = "german_credit_risk.csv"
try:
    df = pd.read_csv(file_name)
except FileNotFoundError:
    print(f"Error: {file_name} not found. ")
```

```python
#Inspecting the data 
print(df.head())
print(df.info())
```

```python
def target_imbalance(df, target_col = 'Risk'):
        counts = df[target_col].value_counts()
        percent = df[target_col].value_counts(normalize = True) * 100
        for label, count in counts.items():
            print(f"Class '{label}': {count} samples ({percent[label]:.2f}%)")
        plt.figure(figsize = (8,5))
        sns.countplot(x = target_col, data = df)
        plt.title('Target Variable Distribution (Imbalance Check)', fontsize=14)
        plt.xlabel('Credit Risk Status', fontsize=12)
        plt.ylabel('Number of Observations', fontsize=12)
        plt.show()
```

```python
target_col = 'Risk'
target_imbalance(df, target_col)
```

```python
#Data Cleaning 
df = df.drop(df.columns[df.columns.str.contains('Unnamed', case=False)], axis=1)
print("Null value numbers")
for col in df:
    count = df[col].isnull().sum()
    if count != 0:
        print(f"{col} has : {count}")
```

```python
#One-Hot Encoding for unranked data 
df = pd.get_dummies(df, columns=['Housing', 'Purpose'], drop_first=True)
```

```python
#Data Preparation
df["Sex_Encoded"] = df["Sex"].map({'female': 0, 'male': 1})
df = df.replace({np.nan: 'none'})
#Ordinal Encoding 
checking_map = {
    'none': 0,
    'little': 1,
    'moderate': 2,
    'rich': 3
}
savings_map = {
    'none': 0,
    'little': 1,
    'moderate': 2,
    'quite rich': 3,
    'rich': 4
}
df['Savings_Encoded'] = df['Saving accounts'].map(savings_map)
df['Checkings_Encoded'] = df['Checking account'].map(checking_map)
```

```python
#Split Data (80% Train, 20% Test) 
y = df['Risk'].map({'good': 0, 'bad': 1})
X = df.drop(['Risk', 'Sex', 'Saving accounts', 'Checking account'], axis = 1)
#Feature Scaling
num_cols = ['Age', 'Credit amount', 'Duration']
cat_cols = [col for col in X.columns if col not in num_cols]
scaler = StandardScaler()
X_scaled_array = scaler.fit_transform(X[num_cols])
X[num_cols] = X_scaled_array
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 50)
X_train
```

```python
#Train Using Logistic Regression 
log_model = LogisticRegression(max_iter = 1000)
log_model.fit(X_train, y_train)
y_pred = log_model.predict(X_test)
```

```python
#Evaluate the Model 
print(f"Accuracy: {accuracy_score(y_test, y_pred):.2f}")
print(classification_report(y_test, y_pred))
```

```python
train_score = log_model.score(X_train, y_train)
test_score = log_model.score(X_test, y_test)
print(f"Training Accuracy: {train_score:.2f}")
print(f"Testing Accuracy: {test_score:.2f}")
```
